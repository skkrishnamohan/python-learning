# Polymorphism

```python
# Polymorphism - 'poly' means many, 'morph' means forms
# Same method name but DIFFERENT behaviour depending on the object

# Real-world analogy:
# 'draw()' means different things for different shapes
#   -> Circle draws a round shape
#   -> Square draws a box
#   -> Triangle draws a triangle
# Same method name 'draw()', but different behaviour for each shape

# In Python, polymorphism is mainly achieved through:
# 1. Method Overriding   - child class redefines a parent's method
# 2. Method Overloading  - NOT supported in Python (shown with error below)
```

### Method Overriding

```python
# Method Overriding - child class provides its OWN version of a parent's method
# The child's version REPLACES the parent's version for that object

class Animal:
    def sound(self):
        print('Animals make sounds')    # generic parent version

class Dog(Animal):
    def sound(self):                    # overriding the parent's sound()
        print('The dog barks')

class Cat(Animal):
    def sound(self):                    # overriding the parent's sound()
        print('The cat meows')

class Cow(Animal):
    def sound(self):                    # overriding the parent's sound()
        print('The cow moos')

a = Animal()
d = Dog()
c = Cat()
w = Cow()

# Same method name, different output per object -> Polymorphism
a.sound()    # Animals make sounds
d.sound()    # The dog barks
c.sound()    # The cat meows
w.sound()    # The cow moos
```

```python
# Polymorphism with a loop
# We can treat different objects uniformly through the common interface

animals = [Dog(), Cat(), Cow()]

for animal in animals:
    animal.sound()    # each object calls its OWN version of sound()
```

```python
# Another example: Shape with area()

class Shape:
    def area(self):
        print('Area of a generic shape')

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        print('Area of Circle:', 3.14 * self.radius * self.radius)

class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width  = width

    def area(self):
        print('Area of Rectangle:', self.length * self.width)

c = Circle(7)
r = Rectangle(5, 4)

c.area()    # Circle's area calculation
r.area()    # Rectangle's area calculation
```

### Method Overloading (NOT supported in Python)

```python
# Method Overloading - same method name, different NUMBER of parameters
# This is supported in Java and C++, but NOT in Python

# In Python, if you define the same method name twice,
# the SECOND definition completely REPLACES the first

class Demo:
    def add(self, a, b):
        print(a + b)

    def add(self, a, b, c):    # this REPLACES the add(a, b) above!
        print(a + b + c)

d = Demo()
d.add(1, 2, 3)    # works fine: 6
d.add(1, 2)       # TypeError: add() missing 1 required positional argument: 'c'
```

# Abstraction

```python
# Abstraction - showing ONLY the essential features, hiding unnecessary details

# Real-world analogy:
# When you press the brake in a car, you don't need to know HOW the braking mechanism works
# You just need to know THAT pressing brake slows down the car
# The internal complexity is HIDDEN from you

# In Python, abstraction is achieved using:
# 1. The 'abc' module (Abstract Base Classes)
# 2. @abstractmethod decorator

# An abstract class:
# - Cannot be instantiated (cannot create objects directly from it)
# - Forces all child classes to implement its abstract methods
```

```python
# Importing ABC (Abstract Base Class) and abstractmethod

from abc import ABC, abstractmethod

# Abstract class - a blueprint that CANNOT be instantiated directly
class Animal(ABC):

    @abstractmethod
    def sound(self):    # abstract method - NO body, child MUST define it
        pass

    def breathe(self):  # concrete method - has a body, inherited as-is
        print('All animals breathe oxygen')

# Child class MUST implement the abstract method
class Dog(Animal):
    def sound(self):
        print('Dog barks')

class Cat(Animal):
    def sound(self):
        print('Cat meows')

d = Dog()
c = Cat()

d.sound()      # Dog barks
d.breathe()    # All animals breathe oxygen
c.sound()      # Cat meows
```

```python
# Trying to create an object of the abstract class directly -> TypeError

from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass

a = Animal()    # TypeError: Can't instantiate abstract class Animal with abstract method sound
```

```python
# If a child class does NOT implement all abstract methods -> also TypeError

from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass

class Fish(Animal):     # Fish does NOT implement sound()
    def swim(self):
        print('Fish swims')

f = Fish()    # TypeError: Can't instantiate abstract class Fish with abstract method sound
```

```python
# Scenario: Payment system using abstraction
# Different payment methods (UPI, Card, Cash) all MUST implement pay()

from abc import ABC, abstractmethod

class Payment(ABC):        # abstract class
    @abstractmethod
    def pay(self, amount): # every payment method must implement this
        pass

class UPI(Payment):
    def pay(self, amount):
        print('Paid Rs', amount, 'via UPI')

class Card(Payment):
    def pay(self, amount):
        print('Paid Rs', amount, 'via Credit Card')

class Cash(Payment):
    def pay(self, amount):
        print('Paid Rs', amount, 'via Cash')

u = UPI()
k = Card()
c = Cash()

u.pay(500)
k.pay(1200)
c.pay(200)
```

```python
# Quick recap of all 4 OOP pillars:

# 1. Encapsulation  - bundling data + methods, using access modifiers to restrict access
#                     Tools: public, _protected, __private, getter/setter

# 2. Inheritance    - child class reuses parent class code
#                     Tools: class Child(Parent), super()

# 3. Polymorphism   - same method name, different behaviour per object
#                     Tools: method overriding in child classes

# 4. Abstraction    - hide implementation details, expose only essential interface
#                     Tools: ABC, @abstractmethod
```
