Rules for creating a variable:

    1. A variable can start with an alphabet or an underscore(_)
    2. Can't start with a number, but it can have a number in the middle or at the end
    3. We can't have special characters in variable(like @,#,$,%,&)
    4. Varaible name can't have space between them(Example: contact num=987654321, it should be contact_num=987654321)
    5. Variables are case sensitive (Ex, Name and name are different)
    6. Can't use keywords as variable names (reserved words that perform some task like for, do, while, if, else, etc.)

# Data Types

```python
#integer - int datatypes, any numeric value, positive or neregative or zero without decimal points

a=10
b=-20
c=0
print(type(a))
print(type(b))
print(type(c))
```

```python
#float - float datatypes, a floating point number is a number that contains the decimal point

d=12.89
e=-67.98
f=.30

print(type(d))
print(type(e))
print(type(f))
```

```python
#complex -A complex number is a number that is written in the form of a + bj or a + bJ.
#Here, 'a' represents the real part of the number and 'b' represents the imaginary part of the number. 
#The suffix 'j' or 'J' after 'b' indicates the square root value of -1. 
#The parts 'a' and 'b' may contain integers or floats. 
#For example, 3+5j, -1-5.5J, 0.2+10.5J are all complex numbers


g=2.5+3.5j
h=3.0-0.5j
i=g+h
print(type(g))
print(type(h))
print(i,'\n',type(i))
```

```python
#str = string datatypes, a string is represented by a group of characters.
#Strings are enclosed wiuth siungle quotes (') or double quotes("). Both are valid.


var1='krishna'
var2="mohan"

# We can also write strings inside the  triple double quotes (""") or triple single quotes('''). Useful to embed a string inside another string.
#Triple‑quoted strings are one of the most flexible features in Python — they let you write multi‑line text, 
#embed quotes, and concatenate variables without escaping everything


var3="My name is " + var1 + var2 + " and I'm Moksha's father."

var4 = """Hello,
My full name is """ + var1 + " " + var2 + """
and I live in Chennai.
"""
print(var3)
print(type(var1))
print(type(var2))
print(type(var3))
print(type(var4))
print(var3,'\n')
print(var4)
```

```python
#bool - Boolean datatypes, There are only two boolean values True or False that can be represented by this datatypes. 
# Python internally represent True as 1 and Flase as 0, a blank string like "" also represented as False.

v1=True
v2=False
#v3=true     # is not valid
print(type(v1))
print(type(v1))
#print(type(v3))

v4='True'
print(type(v4))
#True is 1 and False is 0, 1+0 = 1
v5=v1+v2
print(v5,'\n',type(v5))
```

# Type Casting - Changing of datatypes

```python
#int to float

a= 50
print(type(a))
a=float(a)
print(a, type(a))
```

```python
#float to int

a= 50.02
print(type(a))
a=int(a)
print(a, type(a))
```

```python
# int to str

a= 50.02
print(type(a))
a=str(a)
print(a, type(a))
```

```python
# str to int
# to convert str to int, the variable value should be an integer without a decimal point
# if decimal point is there, we can convert to a float only
a='50'
print(type(a))
a=int(a)
print(a, type(a))

# float to str

a= '50.02'
print(type(a))
a=float(a)
print(a, type(a))
```

```python
#Boolean to int/str
# True is equal to 1
#False is equal to 0

a= True
print(type(a))
a=str(a)
print(a, type(a))
#--------------------------------
a= True
print(type(a))
a=int(a)
print(a, type(a))
#--------------------------------
a= True
print(type(a))
a=float(a)
print(a, type(a))
```

```python
#int to boolean
# Any non-zero value is considered as True. even if it is positive or negative
a= -56.98
print(type(a))
a=bool(a)
print(a, type(a))
```
