# Polymorphism

```python
# Polymorphism - 'poly' means many, 'morph' means forms
# Same method name but DIFFERENT behaviour depending on the object

# Real-world analogy:
# 'draw()' means different things for different shapes
#   -> Circle draws a round shape
#   -> Square draws a box
#   -> Triangle draws a triangle
# Same method name 'draw()', but different behaviour for each shape

# In Python, polymorphism is mainly achieved through:
# 1. Method Overriding   - child class redefines a parent's method
# 2. Method Overloading  - NOT supported in Python (shown with error below)
```

### Method Overriding

```python
# Method Overriding - child class provides its OWN version of a parent's method
# The child's version REPLACES the parent's version for that object

class Animal:
    def sound(self):
        print('Animals make sounds')    # generic parent version

class Dog(Animal):
    def sound(self):                    # overriding the parent's sound()
        print('The dog barks')

class Cat(Animal):
    def sound(self):                    # overriding the parent's sound()
        print('The cat meows')

class Cow(Animal):
    def sound(self):                    # overriding the parent's sound()
        print('The cow moos')

a = Animal()
d = Dog()
c = Cat()
w = Cow()

# Same method name, different output per object -> Polymorphism
a.sound()    # Animals make sounds
d.sound()    # The dog barks
c.sound()    # The cat meows
w.sound()    # The cow moos
```

```python
# Polymorphism with a loop
# We can treat different objects uniformly through the common interface

animals = [Dog(), Cat(), Cow()]

for animal in animals:
    animal.sound()    # each object calls its OWN version of sound()
```

```python
# Another example: Shape with area()

class Shape:
    def area(self):
        print('Area of a generic shape')

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        print('Area of Circle:', 3.14 * self.radius * self.radius)

class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width  = width

    def area(self):
        print('Area of Rectangle:', self.length * self.width)

c = Circle(7)
r = Rectangle(5, 4)

c.area()    # Circle's area calculation
r.area()    # Rectangle's area calculation
```

### Method Overloading (NOT supported in Python)

```python
# Method Overloading - same method name, different NUMBER of parameters
# This is supported in Java and C++, but NOT in Python

# In Python, if you define the same method name twice,
# the SECOND definition completely REPLACES the first

class Demo:
    def add(self, a, b):
        print(a + b)

    def add(self, a, b, c):    # this REPLACES the add(a, b) above!
        print(a + b + c)

d = Demo()
d.add(1, 2, 3)    # works fine: 6
d.add(1, 2)       # TypeError: add() missing 1 required positional argument: 'c'
```
