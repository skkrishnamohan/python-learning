# Operators

```python
# operators are the symbols or keywords that perform operations on value or variables(operands)
# operands     operators     operands
#    3            +             5
```

### Arithmetic Operators

```python
# Used to perform mathamatical calculations
# + - % / ** // *

#Example:

a=10
b=20

print(a+b)  #addition
print(a-b)  #Subraction
print(a*b)  #multiplication
print(a/b)  #Quatient
print(a%b)  #reminder
print(a//b)  #Floor division - returns the quotient rounded down to the nearest whole number
print(a**b)  #addition
```

```python
print(True + False + True + True + False - False)
# When we apply arithmetic operations on the bool  dtype, they behave like int
#       1  +   0   +  1   +  1   +   0   -   0   = 3
```

### Comparison Operators

```python
# Compares two values and returns True or False

num1=int(input("Enter the num1:"))
num2=int(input("Enter the num2:"))

print(num1>num2)  #Greater than
print(num1<num2)  #Lesser than
print(num1==num2)  #Equals to
print(num1!=num2)  #Not equals
print(num1>=num2)  #Greater than or Equal to
print(num1<=num2)  #Lesser than or Equal to
```

### Logical Operators

```python
# Logical operators combine multiple conditions

# and - both Conditions must be true, get true as output
# or - at least one condition must be true
# not - reverse the result

# A   B    A and B     A or B
# T   T      T           T
# T   F      F           T
# F   T      F           T
# F   F      F           F

# A    not A
# T     F
# F     T
```

### Membership operators

```python
# Membership operators: used to check if something exists in a sequence
# in
# not in
# returns boolean values

text='python'
print('P' in text)
print('p' in text)
print('z' not in text)
```

### Identity operators

```python
# Identity operators : check whether two variables refer to the same memory location
# is
# not is
# -5 to 256 - python pre stores them and reuses same memory object for efficiency

#The is and is not operators do not compare the value of the objects. They compare the identity numbers or memory locations of the objects.

a=10
b=10

print(id(a),id(b))
print(a is b)
```

```python
a=256
b=256
print(a is b)
```

```python
a=1000
b=1000
print(a is b)
```

```python
a=6
b=6
c=6
print(a is b is c)
```
