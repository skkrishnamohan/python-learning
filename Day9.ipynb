# Encapsulation

```python
# Encapsulation - wrapping data (attributes) and methods together inside a class
# AND restricting direct access to some of the class's components

# Real-world analogy: A medicine capsule
# You can't see exactly what's inside, but you interact with it through its outer shell

# Why encapsulation?
# 1. Data security - prevents accidental/unauthorised modification
# 2. Controlled access - through getter and setter methods
```

### Access Modifiers

```python
# Python has 3 types of access modifiers:

# 1. Public    - no underscore    (self.name)   -> accessible from anywhere
# 2. Protected - single underscore (self._name) -> should only be used inside class or subclass
# 3. Private   - double underscore (self.__name)-> CANNOT be accessed directly from outside class

class Employee:
    def __init__(self, name, empid, job):
        self.name    = name        # public   - accessible from outside
        self._empid  = empid       # protected - convention: don't access from outside
        self.__job   = job         # private  - CANNOT access directly from outside

e1 = Employee('Rahul', 1001, 'Manager')

print(e1.name)      # works fine - public
print(e1._empid)    # works (but you SHOULD NOT access protected from outside)
```

```python
# Trying to access a private variable directly -> AttributeError

e1 = Employee('Rahul', 1001, 'Manager')
print(e1.__job)     # AttributeError: 'Employee' object has no attribute '__job'
```

### Getter and Setter Methods

```python
# Getter - a method to READ a private variable
# Setter - a method to UPDATE a private variable (with validation if needed)

class Employee:
    def __init__(self, name, empid, job, marks):
        self.name    = name
        self._empid  = empid
        self.__job   = job        # private
        self.__marks = marks      # private

    # Getter - to READ the private __job
    def getter(self):
        print('Job:', self.__job)
        print('Marks:', self.__marks)

    # Setter - to UPDATE the private __job
    def setter(self, new_job):
        self.__job = new_job
        print('Job updated to:', self.__job)

e1 = Employee('Priya', 1002, 'Developer', 90)

e1.getter()              # reading private values via getter
e1.setter('Team Lead')   # updating private value via setter
e1.getter()              # check the updated value
```

# Inheritance

```python
# Inheritance - when one class acquires the properties and methods of another class
# Promotes CODE REUSABILITY - don't rewrite what's already written!

# Real-world analogy: A child inherits traits from parents
# In OOP:  Child class inherits from Parent class

# Terminology:
# Parent class  = Base class  = Super class  (the one being inherited FROM)
# Child class   = Derived class = Sub class  (the one that inherits)

# Types of inheritance:
# 1. Single       - one parent, one child
# 2. Multiple     - two parents, one child
# 3. Multilevel   - A -> B -> C  (chain of inheritance)
# 4. Hierarchical - one parent, multiple children
# 5. Hybrid       - combination of the above types
```

### Single Inheritance

```python
# Single Inheritance: one parent -> one child

# Parent class
class Employee:
    def __init__(self, name, empid):
        self.name  = name
        self.empid = empid

    def display(self):
        print('Name  :', self.name)
        print('EmpID :', self.empid)

# Child class inheriting from Employee
class Tech(Employee):            # Employee is the parent
    def __init__(self, name, empid, language, salary):
        super().__init__(name, empid)    # calling parent's __init__ using super()
        self.language = language
        self.salary   = salary

    def show_tech(self):
        self.display()           # calling parent's method
        print('Language:', self.language)
        print('Salary  :', self.salary)

t1 = Tech('Aman', 2001, 'Python', 80000)
t1.show_tech()
```

```python
# super() - calls the parent class's method
# More elegant than calling Employee.__init__(self, name, empid) directly
# super() automatically knows the parent class

# Child class can:
# 1. Use parent's attributes and methods directly
# 2. Add its own NEW attributes and methods
# 3. Override (change) inherited methods if needed
```

### Hierarchical Inheritance

```python
# Hierarchical: one parent, multiple children

class Animal:
    def __init__(self, name):
        self.name = name

    def eat(self):
        print(self.name, 'is eating')

class Dog(Animal):    # Dog inherits Animal
    def bark(self):
        print(self.name, 'says: Woof!')

class Cat(Animal):    # Cat inherits Animal
    def meow(self):
        print(self.name, 'says: Meow!')

d = Dog('Bruno')
c = Cat('Kitty')

d.eat()     # inherited from Animal
d.bark()    # Dog's own method

c.eat()     # inherited from Animal
c.meow()    # Cat's own method
```

### Multilevel Inheritance

```python
# Multilevel: chain of inheritance  ->  A  ->  B  ->  C
# C inherits from B, and B inherits from A
# C can access methods of BOTH A and B

class Animal:
    def __init__(self, name):
        self.name = name

    def eat(self):
        print(self.name, 'is eating')

class Dog(Animal):       # Dog inherits from Animal
    def bark(self):
        print(self.name, 'says: Woof!')

class Puppy(Dog):        # Puppy inherits from Dog (which already inherits from Animal)
    def play(self):
        print(self.name, 'is playing!')

p = Puppy('Tommy')

p.eat()     # from Animal (grandparent)
p.bark()    # from Dog    (parent)
p.play()    # Puppy's own method
```

### Multiple Inheritance

```python
# Multiple Inheritance: a child class inherits from TWO or more parent classes

class Father:
    def skills_from_father(self):
        print('Cooking, Driving')

class Mother:
    def skills_from_mother(self):
        print('Painting, Singing')

class Child(Father, Mother):    # inheriting from both Father and Mother
    def own_skills(self):
        print('Gaming, Coding')

c = Child()

c.skills_from_father()    # from Father
c.skills_from_mother()    # from Mother
c.own_skills()            # Child's own
```
