# Functions

```python
# Function - a block of reusable code that performs a specific task

# Why functions?
# Without functions: if you need the same logic in 10 places, you write it 10 times
# With functions:    write ONCE, call it anywhere, anytime

# Two types of functions:
# 1. Built-in functions  - already available in Python: print(), len(), type(), input() etc.
# 2. User-defined functions - we create them ourselves using 'def' keyword
```

### Defining and Calling a Function

```python
# Syntax:
# def function_name():
#     statement(s)    <- this is the FUNCTION BODY
#
# function_name()    <- this CALLS the function (executes it)

# Defining the function
def greet():
    print('Hello! Welcome to Python.')
    print('Hope you are enjoying it!')

# Calling the function
greet()    # output: Hello! Welcome to Python.
```

```python
# Functions can be called multiple times!
# This is the main advantage - write once, reuse many times

greet()
greet()
greet()
```

### Parameters and Arguments

```python
# Parameters - variables listed inside the function definition ()
# Arguments  - actual values you pass when calling the function

# Functions with parameters become more flexible

def greet_person(name):       # 'name' is a parameter
    print('Hello,', name, '! Good to see you.')

greet_person('Alice')         # 'Alice' is the argument
greet_person('Bob')
greet_person('Aman')
```

```python
# Multiple parameters - separate with commas

def add(a, b):        # two parameters
    result = a + b
    print(a, '+', b, '=', result)

add(10, 20)           # a=10, b=20
add(5, 7)
add(100, 250)
```

### Return Statement

```python
# return - sends a value BACK to the caller
# Without return, function outputs directly (but can't reuse the result)
# With return, we can store the result in a variable and use it later

def multiply(a, b):
    return a * b        # sends the result back

answer = multiply(6, 7)   # store the returned value
print('Result:', answer)

# Can also use the returned value directly
print('Direct:', multiply(3, 4))
```

```python
# NOTE: if a function has no return statement, it returns None by default

def say_hi():
    print('Hi!')

result = say_hi()
print(result)    # None  <- function printed, but returned nothing
```

```python
# Scenario: BMI Calculator as a function
# BMI = weight / (height * height)

def calculate_bmi(weight, height):
    bmi = weight / (height * height)
    return round(bmi, 2)     # round to 2 decimal places

def interpret_bmi(bmi):
    if bmi < 18.5:
        return 'Underweight'
    elif bmi < 25:
        return 'Normal weight'
    elif bmi < 30:
        return 'Overweight'
    else:
        return 'Obese'

weight = float(input('Enter weight (kg): '))
height = float(input('Enter height (m): '))

bmi = calculate_bmi(weight, height)
category = interpret_bmi(bmi)

print('BMI:', bmi)
print('Category:', category)
```

### Default Parameters

```python
# Default parameters - give a parameter a default value
# If caller doesn't provide an argument, the default is used

def greet_with_message(name, message='Hello'):    # message has a default
    print(message + ',', name + '!')

greet_with_message('Alice')                      # uses default: Hello
greet_with_message('Bob', 'Good morning')        # overrides default
greet_with_message('Charlie', 'Welcome back')
```

### *args (Variable Number of Arguments)

```python
# *args - allows a function to accept ANY number of arguments
# Python packs all extra arguments into a TUPLE

def total(*nums):
    print(nums)         # nums is a tuple of all arguments passed
    print(sum(nums))

total(1, 2, 3)
total(10, 20, 30, 40, 50)
total(5)               # even a single value works

# useful when you don't know how many arguments the caller will pass
```

### Scope of Variables

```python
# Scope - where a variable is accessible

# Local variable  - created inside a function, only accessible inside that function
# Global variable - created outside a function, accessible everywhere

x = 100    # global variable

def show():
    y = 50    # local variable - only exists inside show()
    print('Inside function - x:', x)    # can access global x
    print('Inside function - y:', y)

show()

print('Outside function - x:', x)
```

```python
# Accessing a local variable outside its function -> NameError

def show():
    y = 50

show()
print(y)    # NameError: name 'y' is not defined  (y only lives inside show())
```

### Lambda Functions

```python
# Lambda - a small anonymous (nameless) function written in ONE line
# Used for simple, short operations

# Syntax: lambda parameters : expression

# Normal function:
def square(x):
    return x * x

# Same function as lambda:
sq = lambda x: x * x

print(square(5))    # 25
print(sq(5))        # 25  (same result!)
```

```python
# Lambda with multiple parameters

add = lambda a, b: a + b
print(add(10, 20))    # 30

is_even = lambda n: n % 2 == 0
print(is_even(4))     # True
print(is_even(7))     # False
```
