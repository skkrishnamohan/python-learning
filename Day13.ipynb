## Exercise 1: Class with User Input

```python
# PROBLEM:
# Create a class named Student
# Inside the class, create a function named fun1:
#   -> fun1 should accept user input and RETURN the value
# Create another function named message():
#   -> this function should PRINT the value of the user input

# What this tests:
#   - Creating a class
#   - Using self to share data between methods
#   - input() function inside a class
#   - Return values from methods
```

```python
# SOLUTION:

class Student:
    def fun1(self):
        self.data = input('Enter a word: ')   # store input as an INSTANCE attribute
        return self.data                       # return the value

    def message(self):
        print('You have entered:', self.data)  # access the SAME attribute set by fun1()

# KEY POINT: self.data is what connects fun1() and message()
# fun1() SETS self.data, message() READS self.data
# Without 'self.', the data would be a local variable and message() couldn't see it
```

```python
# Creating object and calling methods

st1 = Student()       # create an object of Student class
st1.fun1()            # calls fun1 -> asks for input -> stores in self.data -> returns it
st1.message()         # calls message -> prints the value stored by fun1()

# IMPORTANT: you must call fun1() BEFORE message()
# If you call message() first, self.data doesn't exist yet -> AttributeError
# This is why __init__ is usually better for setting up data (see Day 8)
```

## Exercise 3: Inheritance + Method Overriding + super()

```python
# PROBLEM:
# Create another class named Modified_Super
# INHERIT from Super class (from Exercise 2)
# Inside Modified_Super:
#   - Create fun1() -> print 'This is function 1 in Modified_Super'
#     Also call the parent's fun1() using super()
#   - Create fun2() -> print 'This is function 2 inside Modified_Super'

# What this tests:
#   - Single inheritance (child inherits from parent)
#   - Method overriding (child redefines fun1())
#   - super() to call parent's version of the method
#   - Adding new methods in child class
```

```python
# SOLUTION:

class Super1:
    def fun1(self):
        print('This is function 1 in Super class')

class Modified_Super1(Super1):        # Modified_Super1 INHERITS from Super1
    def fun1(self):                   # OVERRIDING parent's fun1()
        super().fun1()                # call parent's fun1() FIRST using super()
        print('This is function 1 in Modified_Super')  # then print child's message

    def fun2(self):                   # NEW method — only exists in child
        print('This is function 2 in Modified_Super')
```

```python
obj2 = Modified_Super1()

obj2.fun1()
# Output:
#   This is function 1 in Super class         <- from super().fun1()
#   This is function 1 in Modified_Super       <- from the child's fun1()
# Both messages print because we called super().fun1() INSIDE the override

print()  # blank line for readability

obj2.fun2()
# Output:
#   This is function 2 in Modified_Super
# This is a new method — not overriding anything from parent
```

```python
# KEY TAKEAWAYS from Exercise 3:

# 1. super().fun1() calls the PARENT's version of fun1()
#    Without super(), the parent's fun1() is completely REPLACED

# 2. When you override, you can EXTEND (call parent + add more) or REPLACE (ignore parent)
#    super().fun1() + extra code  = EXTENDING the parent
#    no super() call              = REPLACING the parent

# 3. The child can have methods that the parent doesn't (fun2)
#    But the parent cannot call fun2() — it doesn't know about it
```

## Exercise 4: Abstract Class — Device

```python
# PROBLEM:
# Create an abstract class named Device with an abstract method start()
# Create 2 child classes: Laptop and Mobile
# Both must implement the start() method

# What this tests:
#   - Abstract Base Class (ABC)
#   - @abstractmethod decorator
#   - Child classes MUST implement the abstract method
#   - Adding extra methods in child class
```

```python
# SOLUTION:

from abc import ABC, abstractmethod

class Device(ABC):          # Device is abstract — cannot be instantiated directly
    @abstractmethod
    def start(self):        # abstract method — NO body, child MUST define it
        pass

class Laptop(Device):       # Laptop inherits from Device
    def start(self):        # implementing the abstract method
        print('Laptop started — booting Windows/Linux...')

class Mobile(Device):       # Mobile inherits from Device
    def hello(self):        # extra method — only Mobile has this
        print('Hello from Mobile!')

    def start(self):        # implementing the abstract method
        print('Mobile started — loading Android/iOS...')
```

```python
# Creating objects and calling methods

lap = Laptop()
lap.start()        # Laptop started — booting Windows/Linux...

mob = Mobile()
mob.start()        # Mobile started — loading Android/iOS...
mob.hello()        # Hello from Mobile!  (extra method, not from Device)

# What if Mobile didn't implement start()?
# -> TypeError: Can't instantiate abstract class Mobile with abstract method start
# The ABC FORCES you to implement all abstract methods
```

```python
# Can we create an object of Device directly?

# d = Device()   # TypeError: Can't instantiate abstract class Device

# NO! Device is just a BLUEPRINT
# It says 'every device must have a start() method'
# But it doesn't define WHAT start() does — that's the child's job
```

## Exercise 5: Abstract Class — Shape with area()

```python
# PROBLEM:
# Create an abstract class Shape with an abstract method area()
# Create two child classes: Circle and Rectangle
# Each must implement area() with the correct formula

# What this tests:
#   - Abstraction + __init__ together
#   - Each child has DIFFERENT attributes (radius vs length/width)
#   - Same method name area(), different formulas -> polymorphism!
```

```python
# SOLUTION:

from abc import ABC, abstractmethod

class Shape(ABC):                   # abstract parent
    @abstractmethod
    def area(self):                 # every shape must be able to calculate its area
        pass

class Circle(Shape):                # inherits from Shape
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        print('Area of the circle is:', 3.14 * self.radius ** 2)

class Rectangle(Shape):             # inherits from Shape
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        print('Area of the rectangle is:', self.length * self.width)
```

```python
c1 = Circle(5)
c1.area()              # Area of the circle is: 78.5

r1 = Rectangle(10, 5)
r1.area()              # Area of the rectangle is: 50

# Both shapes have area() — but the FORMULA is different
# Circle uses π * r²
# Rectangle uses length * width
# This is abstraction (blueprint) + polymorphism (same interface, different behaviour)
```

## Exercise 6: Multiple Inheritance — Teacher & Mentor

```python
# PROBLEM:
# Demonstrate multiple inheritance with:
#   - Teacher class with a show() method
#   - Mentor class with a show() method
#   - Guide class that INHERITS from BOTH Teacher and Mentor
#   - Guide should be able to call show() from both parents

# What this tests:
#   - Multiple inheritance (one child, two parents)
#   - How to call specific parent methods when both have same name
#   - ClassName.method(self) syntax vs super()
```

```python
# SOLUTION:

class Teacher:
    def show(self):
        print('I am a Teacher')

class Mentor:
    def show(self):
        print('I am a Mentor')

class Guide(Teacher, Mentor):       # Guide inherits from BOTH Teacher and Mentor
    def show(self):
        print('I am a Guide')        # child's own version
        Teacher.show(self)            # explicitly call Teacher's show()
        Mentor.show(self)             # explicitly call Mentor's show()
```

```python
g = Guide()
g.show()

# Output:
#   I am a Guide      <- Guide's own show()
#   I am a Teacher     <- Teacher's show() called explicitly
#   I am a Mentor      <- Mentor's show() called explicitly
```

```python
# WHY use Teacher.show(self) instead of super().show()?

# When BOTH parents have the SAME method name, super() follows MRO
# (Method Resolution Order) and calls only ONE parent's version

# super().show() here would call ONLY Teacher's show() (first parent in list)
# It would NOT call Mentor's show()

# To call BOTH parents' versions, we use:
#   Teacher.show(self)   <- explicitly call Teacher's version
#   Mentor.show(self)    <- explicitly call Mentor's version

# You can check the MRO with:
print(Guide.__mro__)
# (<class 'Guide'>, <class 'Teacher'>, <class 'Mentor'>, <class 'object'>)
```

## Exercise 7: Abstract Class — Ride Fare Calculator

```python
# PROBLEM:
# Create an abstract class Ride with an abstract method calculate_fare()
# Create 2 child classes: Bus and Car
# Bus: per kilometer fare = Rs 5
# Car: per kilometer fare = Rs 20
# Both accept distance in __init__

# What this tests:
#   - Abstraction with __init__
#   - Different implementations of the same abstract method
#   - Using instance attributes in the calculation
```

```python
# SOLUTION:

from abc import ABC, abstractmethod

class Ride(ABC):
    @abstractmethod
    def calculate_fare(self):
        pass

class Bus(Ride):
    def __init__(self, distance):
        self.distance = distance

    def calculate_fare(self):
        fare = 5 * self.distance
        print(f'Bus fare for {self.distance} km: Rs {fare}')

class Car(Ride):
    def __init__(self, distance):
        self.distance = distance

    def calculate_fare(self):
        fare = 20 * self.distance
        print(f'Car fare for {self.distance} km: Rs {fare}')
```

```python
bus = Bus(10)
bus.calculate_fare()    # Bus fare for 10 km: Rs 50

car = Car(20)
car.calculate_fare()    # Car fare for 20 km: Rs 400

# Same method name calculate_fare(), but different rates
# Bus charges Rs 5/km, Car charges Rs 20/km
# The abstract class GUARANTEES both will have calculate_fare()
```
