# Loops

```python
# A loop is a control statement that allows a block of code to be executed REPEATEDLY
# Without loops, we would have to write the same code many times

# Example without a loop:
# print(1)
# print(2)
# print(3)  ... imagine doing this for 1 to 100!

# Two types of loops in Python:
# for   - when we KNOW how many times to loop
# while - when we loop UNTIL a condition becomes False
```

### for loop

```python
# for loop - used when we know the number of iterations
# range(start, stop) -> generates numbers from start up to stop-1  (n-1 rule)

# Syntax:
# for variable in range(start, stop):
#     statement(s)

for i in range(1, 6):    # prints 1, 2, 3, 4, 5 (stops before 6)
    print(i)
```

```python
# range(stop)         -> starts from 0 by default
# range(start, stop)  -> start to stop-1
# range(start, stop, step) -> step controls how much to jump each time

for i in range(1, 11, 2):    # 1, 3, 5, 7, 9  (jumps by 2)
    print(i)
```

```python
# Scenario: Print multiplication table of a number

num = int(input('Enter a number for multiplication table: '))

for i in range(1, 11):
    print(num, 'x', i, '=', num * i)
```

```python
# Scenario: Find the sum of first n natural numbers

n = int(input('Enter n: '))
total = 0

for i in range(1, n + 1):    # 1 to n (inclusive)
    total = total + i        # add each number to total

print('Sum of first', n, 'natural numbers is:', total)
```

### while loop

```python
# while loop - works until the condition is True
# Use when we DON'T know the exact number of repetitions

# Syntax:
# while condition:
#     statement(s)
#     increment/decrement  <- IMPORTANT: without this, loop runs forever (infinite loop)!

n = 1
while n <= 10:
    print(n)
    n += 1    # increment factor - n = n + 1
```

```python
# Scenario: Keep asking user for input until they type 'quit'

user_input = ''
while user_input != 'quit':
    user_input = input('Type something (type quit to exit): ')
    print('You typed:', user_input)
```

```python
# Nested for loop - a loop inside another loop
# Outer loop runs once -> inner loop runs COMPLETELY -> outer loop runs again

# Example: multiplication table for 1 to 3
for i in range(1, 4):          # outer loop  (tables 1, 2, 3)
    for j in range(1, 6):      # inner loop  (1 to 5)
        print(i, 'x', j, '=', i * j)
    print()                    # blank line after each table
```

### break and continue

```python
# continue - skips the current iteration and moves to the next one
# The loop does NOT stop, it just skips this particular round

# Example: print 1 to 10 but skip 5

for i in range(1, 11):
    if i == 5:      # when i is 5, skip this iteration
        continue
    print(i)
```

```python
# break - stops/terminates the entire loop immediately

# Example: print numbers 1 to 10, but stop when we reach 5

for i in range(1, 11):
    print(i)
    if i == 5:
        break        # stop the loop here
```

```python
# Scenario: Find the factorial of a number
# Factorial of 5 = 5 x 4 x 3 x 2 x 1 = 120

n = int(input('Enter a number: '))
factorial = 1

for i in range(1, n + 1):
    factorial = factorial * i

print('Factorial of', n, 'is:', factorial)
```

```python
# Scenario: Print a star pattern
# *
# * *
# * * *
# * * * *
# * * * * *

n = 5
for i in range(1, n + 1):
    print('* ' * i)     # repeat '* ' i times
```

```python
# Scenario: Check whether a number is prime or not
# A prime number is divisible ONLY by 1 and itself

num = int(input('Enter a number: '))
is_prime = True

if num < 2:
    is_prime = False
else:
    for i in range(2, num):      # check divisibility from 2 to num-1
        if num % i == 0:
            is_prime = False
            break                # no need to check further

if is_prime:
    print(num, 'is a prime number')
else:
    print(num, 'is NOT a prime number')
```

# List

```python
# List [] - a collection of elements stored in a sequential manner

# Properties of a List:
# 1. Ordered    - elements maintain their order
# 2. Mutable    - can be changed after creation (add, remove, update)
# 3. Duplicates allowed - same value can appear multiple times
# 4. Mixed types - can store int, float, str, bool all together

sports = ['Football', 'Badminton', 'Cricket', 'Hockey', 'Basketball', 'Tennis', 'Golf']
print(sports)
print(type(sports))
```

```python
# Lists can store mixed data types and duplicates

mixed = [1, 'Bob', 15, 15, 15, 8.9, True]
print(mixed)
```

### Indexing and Slicing

```python
# Indexing - accessing a particular element from the list

# Positive Indexing: starts from 0, left to right
# Negative Indexing: starts from -1, right to left

# sports = ['Football', 'Badminton', 'Cricket', 'Hockey', 'Basketball', 'Tennis', 'Golf']
#  index:        0           1           2          3           4           5        6
#  neg index:   -7          -6          -5         -4          -3          -2       -1

sports = ['Football', 'Badminton', 'Cricket', 'Hockey', 'Basketball', 'Tennis', 'Golf']

print(sports[0])     # First element
print(sports[3])     # Fourth element
print(sports[-1])    # Last element
print(sports[-4])    # 4th from the end
print(len(sports))   # Total number of elements
```

```python
# Slicing - accessing multiple elements at once
# Syntax: list[start : stop : step]   <- n-1 rule applies for stop

print(sports[1:4])    # index 1, 2, 3  (4-1 = 3, so stops at index 3)
print(sports[2:6])    # index 2, 3, 4, 5
print(sports[:5])     # from start up to index 4
print(sports[:])      # all elements
print(sports[::2])    # every 2nd element (step=2)
print(sports[::-1])   # reverse the list (step=-1)
```

### List Methods

```python
# append() - adds a single element at the END of the list
# insert()  - adds element at a SPECIFIC index
# pop()     - removes element using its INDEX (returns the removed element)
# remove()  - removes element using its VALUE
# reverse() - reverses the order of the list
# extend()  - joins another list to the current list
# clear()   - empties the entire list
# index()   - returns the index position of a value

l = [10, 20, 30, 40, 50]
print('Original:', l)
```

```python
l.append(60)           # adds 60 at the end
print('After append:', l)

l.insert(2, 25)        # inserts 25 at index 2
print('After insert:', l)
```

```python
l.pop(3)               # removes element at index 3
print('After pop:', l)

l.remove(50)           # removes the value 50
print('After remove:', l)
```

```python
lis1 = [10, 20, 30]
lis2 = [90, 100]

lis1.extend(lis2)      # joins lis2 to the end of lis1
print('After extend:', lis1)

lis1.reverse()         # reverses in place
print('After reverse:', lis1)
```

```python
# Iterating through a list using for loop

fruits = ['apple', 'banana', 'mango', 'orange']

for fruit in fruits:
    print(fruit)
```

```python
# enumerate() - gives both the INDEX and the VALUE when iterating a list
# useful when you need to know the position of each element

fruits = ['apple', 'banana', 'mango', 'orange']

for index, fruit in enumerate(fruits):
    print(index, '-', fruit)

# by default index starts at 0, can change with start parameter
print()
for index, fruit in enumerate(fruits, start=1):    # start numbering from 1
    print(index, '.', fruit)
```

# Tuple

```python
# Tuple () - exactly like a List BUT it is IMMUTABLE (cannot be changed after creation)

# Use Tuple when the data should NOT be changed
# Example: days of the week, months, coordinates

# Properties:
# 1. Ordered
# 2. IMMUTABLE - cannot add, remove or change elements
# 3. Duplicates allowed
# 4. Mixed types allowed

days = ('Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday')
print(days)
print(type(days))
```

```python
# Indexing and slicing works exactly the same as lists

a = (1, 2, 3, 4, 'ab', 8.9)

print(a[0])      # first element
print(a[-1])     # last element
print(a[1:4])    # slicing
```

```python
# This will throw a TypeError because tuples are immutable
# Once created, you CANNOT change any element

a = (1, 2, 3, 4, 'ab')
a[2] = 100    # TypeError: 'tuple' object does not support item assignment
```
