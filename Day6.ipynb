# Set

```python
# Set {} - an unordered collection that does NOT allow duplicate values

# Properties:
# 1. UNORDERED    - no fixed position, order can change on each run
# 2. No DUPLICATES - duplicate values are automatically removed
# 3. Mutable      - can add and remove elements
# 4. NO Indexing  - cannot access elements by index (no order = no index)

a = {'apple', 1, 'mango', 'mango', 5, 9}    # 'mango' appears twice
print(a)      # notice: 'mango' will appear only ONCE
print(type(a))
```

```python
# Interesting: True = 1 and False = 0 in Python
# So if 1 and True both exist, the set keeps only ONE (they are duplicates!)

s = {1, True, 0, False, 2, 3}
print(s)    # output: {0, 1, 2, 3}  -- True removed (same as 1), False removed (same as 0)
```

```python
# Indexing does NOT work in sets - will throw a TypeError

a = {'apple', 1, 'mango', 5, 9}
print(a[0])    # TypeError: 'set' object is not subscriptable
```

### Set Methods

```python
# add()    - adds a single element to the set
# remove() - removes an element (raises error if element not found)
# discard()- removes an element (NO error even if element not found)
# clear()  - empties the entire set
# len()    - returns number of elements

a = {'apple', 1, 'mango', 5, 9}
print('Original:', a)

a.add(10)
print('After add(10):', a)

a.remove(5)
print('After remove(5):', a)

a.discard(999)    # 999 is not in the set, but no error!
print('After discard(999):', a)
```

### Set Operations (like Venn Diagrams)

```python
# Set operations come from mathematics (Venn Diagrams)
# union()        - all elements from BOTH sets (no duplicates)
# intersection() - ONLY elements that exist in BOTH sets
# difference()   - elements in set A but NOT in set B

a = {1, 2, 3, 4, 5}
b = {4, 5, 6, 7, 8}

print('Set A:', a)
print('Set B:', b)
print()

print('union (A | B):', a.union(b))                # all elements from both
print('intersection (A & B):', a.intersection(b))  # only common elements
print('difference (A - B):', a.difference(b))      # in A but not in B
```

```python
# Scenario: Find common interests between two friends

alice_hobbies = {'reading', 'gaming', 'cooking', 'cycling'}
bob_hobbies   = {'gaming', 'cycling', 'painting', 'hiking'}

common = alice_hobbies.intersection(bob_hobbies)
print('Common hobbies:', common)

all_hobbies = alice_hobbies.union(bob_hobbies)
print('All hobbies combined:', all_hobbies)

only_alice = alice_hobbies.difference(bob_hobbies)
print('Only Alice enjoys:', only_alice)
```

# Dictionary

```python
# Dictionary {} - stores data as KEY : VALUE pairs
# Think of it like a real dictionary: word (key) -> meaning (value)
# Or like a contact book: name (key) -> phone number (value)

# Properties:
# 1. Keys must be UNIQUE - duplicate keys are not allowed
# 2. Values can be duplicated
# 3. Ordered (from Python 3.7+)
# 4. Mutable - can add, update, delete

# Syntax: {key1: value1, key2: value2, ...}

player = {'Sehwag': 89, 'Sachin': 100, 'Dravid': 50, 'Ganguly': 70}
print(player)
print(type(player))
```

```python
# Accessing values using keys
# NOT by index! (unlike list)

print(player['Sachin'])    # 100
print(player['Dravid'])    # 50
```

```python
# If we try to access a key that doesn't exist -> KeyError

print(player['Kohli'])     # KeyError: 'Kohli'
```

### Dictionary Methods

```python
# keys()   - returns all the keys
# values() - returns all the values
# items()  - returns all key-value pairs
# get()    - access a value safely (no KeyError if key missing, returns None)
# update() - add or update key-value pairs
# pop()    - removes a key-value pair by key
# clear()  - empties the dictionary

player = {'Sehwag': 89, 'Sachin': 100, 'Dravid': 50, 'Ganguly': 70}

print('Keys:', player.keys())
print('Values:', player.values())
print('Items:', player.items())
```

```python
# get() - safer way to access values
# Returns None (or a default) if key doesn't exist, no error!

print(player.get('Sachin'))     # 100
print(player.get('Kohli'))      # None (no error)
print(player.get('Kohli', 0))   # 0 (custom default value)
```

```python
# update() and pop()

player.update({'Kohli': 95})        # add new key
print('After update:', player)

player.update({'Sachin': 110})      # update existing key
print('Sachin updated:', player)

player.pop('Sehwag')                # remove Sehwag
print('After pop Sehwag:', player)
```

```python
# Iterating through a dictionary

student = {'name': 'Aman', 'age': 20, 'grade': 'A', 'city': 'Delhi'}

# Loop through keys
for key in student:
    print(key, '->', student[key])
```

```python
# Scenario: Simple phone book

contacts = {
    'Mum':    '9876543210',
    'Dad':    '9123456789',
    'Friend': '9988776655'
}

name = input('Enter name to search: ')

number = contacts.get(name)

if number:
    print('Phone number of', name, 'is:', number)
else:
    print(name, 'is not in contacts')
```

```python
# Quick comparison: List vs Tuple vs Set vs Dictionary

# List       []  - ordered, mutable,   duplicates allowed, indexed by position
# Tuple      ()  - ordered, immutable, duplicates allowed, indexed by position
# Set        {}  - unordered, mutable, NO duplicates,      NO indexing
# Dictionary {}  - ordered, mutable,   unique KEYS,        indexed by key

# when to use which:
# List  -> when order matters and you'll be adding/removing elements
# Tuple -> when data should not change (days of week, coordinates, settings)
# Set   -> when you only care about unique values (remove duplicates quickly)
# Dict  -> when you want to label your data with meaningful names (key-value)
```
