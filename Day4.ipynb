# Conditional Statements

```python
# Conditional Statements allow the program to make decisions
# and execute different blocks of code based on conditions

# Real-world analogy - Traffic light:
#     If  light is red     -> Stop
#     elif light is yellow  -> Slow down
#     else (green)          -> Go

# Keywords: if , elif , else
```

### if statement

```python
# Syntax:
# if condition:
#     statement(s)    <- indentation (4 spaces) tells Python what belongs inside the if block

# The code inside the if block runs ONLY when the condition is True

age = 20
if age >= 18:
    print('Eligible to vote')
```

### if-else statement

```python
# if-else: two paths
# condition is True  -> if block runs
# condition is False -> else block runs (default)

num = int(input('Enter a number: '))
if num > 0:
    print('Positive number')
else:
    print('Negative number or zero')
```

```python
# Scenario: Check if a person is eligible to vote in India

age = int(input('Enter your age: '))

if age >= 18:
    print('You are eligible to vote in India.')
else:
    print('You are not eligible to vote in India.')
```

```python
# Scenario: Check if a number is even or odd
# A number is even if it is divisible by 2 (remainder = 0)

num = int(input('Enter a number: '))
if num % 2 == 0:    # % gives the remainder
    print('Even number')
else:
    print('Odd number')
```

### if-elif-else

```python
# if-elif-else is used when we have MORE than 2 conditions to check
# elif = else if  -> checks another condition if the previous one was False
# Only ONE block runs -> the FIRST condition that is True
# else at the end is the default -> runs when NO condition above is True

a = int(input('Enter the value of a: '))
b = int(input('Enter the value of b: '))

if a > b:
    print('a is the greater number')
elif a == b:
    print('Both values are equal')
else:                        # default - none of the above was True
    print('b is the greater number')
```

```python
# Scenario: Grade Calculator
# Assign grades: O, A, B, C, D or Fail based on marks

marks = int(input('Enter your marks: '))

if marks > 90:
    print('Grade: O')
elif marks >= 80 and marks <= 90:
    print('Grade: A')
elif marks >= 70 and marks < 80:
    print('Grade: B')
elif marks >= 60 and marks < 70:
    print('Grade: C')
elif marks >= 40 and marks < 60:
    print('Grade: D')
else:
    print('Fail')
```

```python
# Scenario: Check if a number is positive, negative, or zero

num = int(input('Enter a number: '))

if num > 0:
    print('Positive number')
elif num < 0:
    print('Negative number')
else:
    print('Zero')
```

### Nested if

```python
# Nested if - a condition INSIDE another condition (if inside if)
# Use when the second check DEPENDS on the first one being True

# Difference:
# if-elif   -> Independent conditions (any one can be True on its own)
# Nested if -> Dependent conditions (inner if only runs when outer if is True)
```

```python
# Scenario: Voting eligibility with Voter ID check
# Step 1: Check age (>= 18)
# Step 2: If age is valid, THEN check if the person has a Voter ID

age = int(input('Enter your age: '))

if age >= 18:
    voter_id = input('Do you have a Voter ID? (yes/no): ')
    if voter_id.lower() == 'yes':     # .lower() converts 'Yes', 'YES' etc to 'yes'
        print('Eligible to vote')
    else:
        print('Not eligible - no Voter ID')
else:
    print('Not eligible - age below 18')
```

```python
# Scenario: Find the greatest of three numbers

a = int(input('Enter a: '))
b = int(input('Enter b: '))
c = int(input('Enter c: '))

if a > b:
    if a > c:
        print('a is the greatest:', a)
    else:
        print('c is the greatest:', c)
else:
    if b > c:
        print('b is the greatest:', b)
    else:
        print('c is the greatest:', c)
```

### Shorthand if (Ternary Operator)

```python
# Shorthand if / Ternary operator - write an if-else in ONE LINE
# Syntax: value_if_true  if  condition  else  value_if_false

num = int(input('Enter a number: '))
result = 'Even' if num % 2 == 0 else 'Odd'
print(result)

# without shorthand (same logic, more lines):
# if num % 2 == 0:
#     result = 'Even'
# else:
#     result = 'Odd'

age = int(input('Enter age: '))
status = 'Adult' if age >= 18 else 'Minor'
print(status)
```
