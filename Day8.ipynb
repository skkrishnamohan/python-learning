# Object Oriented Programming (OOP)

```python
# OOP - Object Oriented Programming
# A way of writing programs by modelling real-world things as 'objects'

# Real-world example: A Car
# Attributes (properties): name, color, model, price, fuel type
# Methods (actions/behaviour): start(), stop(), accelerate(), brake()

# In OOP:
# Class  - the BLUEPRINT / template (like a car design on paper)
# Object - the ACTUAL THING created from the blueprint (the real car in your garage)

# One blueprint (class) can create many objects (instances)
# e.g. Same Car class -> Thar object, Swift object, BMW object
```

```python
# More real-world examples:

# Class: Animal
#   Attributes: name, species, age, sound
#   Methods: eat(), sleep(), makeSound()

# Class: BankAccount
#   Attributes: account_number, owner_name, balance
#   Methods: deposit(), withdraw(), check_balance()

# Class: Student
#   Attributes: name, roll_number, marks
#   Methods: study(), appear_exam(), get_grade()
```

### Creating a Class

```python
# Syntax to create a class:
# class ClassName:      <- class name should start with Capital letter
#     def __init__(self, param1, param2, ...):
#         self.attribute1 = param1
#         self.attribute2 = param2
#
#     def method_name(self):
#         statement(s)

class Car:
    # __init__ is called automatically when we create an object
    # It initialises the object's attributes
    def __init__(self, name, color, model, price):
        self.name  = name       # self.attribute = parameter
        self.color = color
        self.model = model
        self.price = price

    # A method to display car details
    def display(self):
        print('Car Name :', self.name)
        print('Color    :', self.color)
        print('Model    :', self.model)
        print('Price    :', self.price)
```

### Creating Objects from a Class

```python
# Creating objects (instances) from the Car class
# Syntax: object_name = ClassName(argument1, argument2, ...)

ob1 = Car('Thar', 'Black', 2026, 2000000)    # creating first car object
ob2 = Car('Swift', 'White', 2024, 700000)    # creating second car object
ob3 = Car('BMW', 'Blue', 2025, 8500000)      # creating third car object

# Calling the display method on each object
ob1.display()
print()
ob2.display()
print()
ob3.display()
```

```python
# Accessing attributes directly using dot notation

print('First car name:', ob1.name)
print('Second car price:', ob2.price)
print('Third car color:', ob3.color)
```

### Understanding self and __init__

```python
# self - refers to the CURRENT object being created/used
# It is automatically passed when you call a method on an object
# ob1.display()  ->  Python internally calls display(ob1)

# __init__ - the CONSTRUCTOR method
# Called automatically when you create an object with ClassName()
# Used to SET UP the initial values of the object

class Dog:
    def __init__(self, name, breed, age):
        self.name  = name
        self.breed = breed
        self.age   = age

    def bark(self):
        print(self.name, 'says: Woof!')

    def info(self):
        print(self.name, 'is a', self.breed, 'aged', self.age, 'years')

d1 = Dog('Bruno', 'Labrador', 3)
d2 = Dog('Max', 'Poodle', 1)

d1.bark()
d1.info()
d2.bark()
d2.info()
```

```python
# Scenario: Create a Student class

class Student:
    def __init__(self, name, roll_no, marks):
        self.name    = name
        self.roll_no = roll_no
        self.marks   = marks

    def display(self):
        print('Name    :', self.name)
        print('Roll No :', self.roll_no)
        print('Marks   :', self.marks)

    def get_grade(self):
        if self.marks >= 90:
            return 'O'
        elif self.marks >= 80:
            return 'A'
        elif self.marks >= 70:
            return 'B'
        elif self.marks >= 60:
            return 'C'
        else:
            return 'Fail'

s1 = Student('Aman', 101, 88)
s2 = Student('Priya', 102, 55)

s1.display()
print('Grade:', s1.get_grade())
print()
s2.display()
print('Grade:', s2.get_grade())
```
