## Exercise 1: Ride Booking System (Abstraction)

```python
# PROBLEM:
# Design a Ride Booking System:
#   - Abstract class 'Ride' with abstract method calculate_fare(distance)
#   - Create two concrete classes: BikeRide and CarRide
#   - BikeRide charges Rs 25 per km
#   - CarRide charges Rs 40 per km
#   - Both inherit from Ride

# What this tests:
#   - Abstract class as a contract (blueprint)
#   - distance is passed as a PARAMETER to the method (not __init__)
#   - Same interface, different pricing logic
```

```python
# SOLUTION:

from abc import ABC, abstractmethod

class Ride(ABC):
    @abstractmethod
    def calculate_fare(self, distance):    # distance is a method parameter here
        pass

class BikeRide(Ride):
    def calculate_fare(self, distance):
        fare = distance * 25
        print(f'Bike fare for {distance} km: Rs {fare}')

class CarRide(Ride):
    def calculate_fare(self, distance):
        fare = distance * 40
        print(f'Car fare for {distance} km: Rs {fare}')
```

```python
b = BikeRide()
c = CarRide()

b.calculate_fare(30)    # Bike fare for 30 km: Rs 750
c.calculate_fare(30)    # Car fare for 30 km: Rs 1200

# Notice: distance is NOT stored in __init__ here
# Instead, it's passed each time you call calculate_fare()
# This makes the objects reusable — same object, different distances
```

```python
# COMPARE: Day 13 Exercise 7 vs Day 14 Exercise 1

# Day 13 approach: distance in __init__
#   Bus(10).calculate_fare()    <- distance set once at creation
#   Need new object for different distance

# Day 14 approach: distance as method parameter
#   BikeRide().calculate_fare(10)   <- distance passed each time
#   BikeRide().calculate_fare(50)   <- same object, different distance

# Both are valid! It depends on what makes sense for your use case
# If distance is fixed for a ride -> __init__ is better
# If you want to calculate fares for multiple distances -> parameter is better
```

## Exercise 2: Teacher Inheritance

```python
# PROBLEM:
# Create a class Teacher with a method teach()
# Inherit it into MathTeacher and ScienceTeacher
# Show how each subclass gets access to the parent's method

# What this tests:
#   - Simple single inheritance
#   - Child classes inherit parent methods automatically
#   - Each child can also have its OWN unique methods
```

```python
# SOLUTION:

class Teacher:
    def teach(self):
        print('Teacher teaches us')      # common behaviour for all teachers

class MathTeacher(Teacher):              # inherits from Teacher
    def math(self):                      # own method — specific to MathTeacher
        print('I teach Mathematics')

class ScienceTeacher(Teacher):           # inherits from Teacher
    def science(self):                   # own method — specific to ScienceTeacher
        print('I teach Science')
```

```python
m = MathTeacher()
s = ScienceTeacher()

m.teach()      # Teacher teaches us       <- inherited from Teacher
m.math()       # I teach Mathematics      <- MathTeacher's own method

print()  # blank line

s.teach()      # Teacher teaches us       <- inherited from Teacher
s.science()    # I teach Science           <- ScienceTeacher's own method

# KEY POINT: Both children get teach() for FREE from the parent
# They didn't have to write teach() again — inheritance gives it to them
# But each child also has its own unique method (math() or science())
```

```python
# This is called HIERARCHICAL INHERITANCE
# One parent -> Multiple children
#
#       Teacher
#      /       \
# MathTeacher  ScienceTeacher
#
# We covered this pattern in Day 9
```

## Exercise 3: Student Result System (Class with Logic)

```python
# PROBLEM:
# Create a class Student with attributes: name, rollno, marks
# Add a method grade() that decides the grade based on marks:
#   marks >= 90 -> Grade A
#   marks >= 75 -> Grade B
#   marks >= 60 -> Grade C
#   else        -> Debarred
# Create objects and test the grading system

# What this tests:
#   - __init__ with multiple attributes
#   - Conditional logic (if-elif-else) inside a method
#   - Using self to access instance attributes in methods
```

```python
# SOLUTION:

class Student:
    def __init__(self, name, rollno, marks):
        self.name = name
        self.rollno = rollno
        self.marks = marks

    def grade(self):
        if self.marks >= 90:
            print(f'{self.name} (Roll No: {self.rollno}) -> Grade A')
        elif self.marks >= 75:
            print(f'{self.name} (Roll No: {self.rollno}) -> Grade B')
        elif self.marks >= 60:
            print(f'{self.name} (Roll No: {self.rollno}) -> Grade C')
        else:
            print(f'{self.name} (Roll No: {self.rollno}) -> Debarred')
```

```python
s1 = Student('Shriti', 6, 54)
s1.grade()     # Shriti (Roll No: 6) -> Debarred

s2 = Student('Vinay', 7, 78)
s2.grade()     # Vinay (Roll No: 7) -> Grade B

s3 = Student('Priya', 3, 95)
s3.grade()     # Priya (Roll No: 3) -> Grade A

# Each Student object stores its OWN name, rollno, and marks
# The grade() method uses self.marks to decide the grade
# Same method, different result per student — that's OOP!
```

## Exercise 4: Polymorphism — Animal speak()

```python
# PROBLEM:
# Create a base class Animal with method speak()
# Inherit it in Dog and Cat classes
# Override the speak() method in each child class

# What this tests:
#   - Method overriding (child replaces parent's method)
#   - Polymorphism (same method name, different behaviour)
```

```python
# SOLUTION:

class Animal:
    def speak(self):
        print('Animals can speak!')     # generic parent version

class Dog(Animal):
    def speak(self):                    # OVERRIDES Animal's speak()
        print('Dog says: Bark!!')

class Cat(Animal):
    def speak(self):                    # OVERRIDES Animal's speak()
        print('Cat says: Meow!')
```

```python
d = Dog()
c = Cat()

d.speak()    # Dog says: Bark!!      <- Dog's version, NOT Animal's
c.speak()    # Cat says: Meow!       <- Cat's version, NOT Animal's

# Even though Dog and Cat inherit from Animal,
# they have their OWN version of speak()
# Python uses the CHILD's version when available — this is method overriding
```

```python
# Polymorphism with a loop — treating different objects uniformly

animals = [Dog(), Cat(), Dog(), Cat()]

for animal in animals:
    animal.speak()    # each object calls its OWN version of speak()

# Output:
#   Dog says: Bark!!
#   Cat says: Meow!
#   Dog says: Bark!!
#   Cat says: Meow!

# This is the POWER of polymorphism:
# The loop doesn't know (or care) what TYPE of animal it is
# It just calls speak() and the RIGHT version runs automatically
```

## Exercise 5: Food Delivery App (Multiple Inheritance)

```python
# PROBLEM:
# You are developing a food delivery app
# Customer class: stores customer name, has place_order(item) method
# Location class: stores address, has show_address() method
# OrderDetails class: inherits BOTH Customer and Location
#   -> combines ordering + delivery address in one class

# What this tests:
#   - Multiple inheritance with __init__ in both parents
#   - Calling each parent's __init__ explicitly
#   - A child that combines functionality from two unrelated parents
```

```python
# SOLUTION:

class Customer:
    def __init__(self, name):
        self.name = name

    def place_order(self, item):
        print(f'{self.name} has ordered {item}')

class Location:
    def __init__(self, address):
        self.address = address

    def show_address(self):
        print(f'Delivery address: {self.address}')

class OrderDetails(Customer, Location):     # inherits from BOTH
    def __init__(self, name, address):
        Customer.__init__(self, name)        # call Customer's __init__
        Location.__init__(self, address)      # call Location's __init__
```

```python
order = OrderDetails('Ravi', 'BTM Layout, 5th Cross')
order.place_order('Burger')    # Ravi has ordered Burger     <- from Customer
order.show_address()           # Delivery address: BTM Layout, 5th Cross  <- from Location

# OrderDetails COMBINES Customer + Location
# It has access to methods from BOTH parents
# This is a practical use of multiple inheritance
```

```python
# KEY POINT: Why Customer.__init__(self, name) instead of super().__init__()?

# When you have multiple inheritance with __init__ in BOTH parents:
#   super().__init__() will only call the FIRST parent's __init__ (by MRO)
#   It will NOT automatically call BOTH parents

# So we explicitly call each parent's __init__:
#   Customer.__init__(self, name)     <- sets self.name
#   Location.__init__(self, address)  <- sets self.address

# This ensures BOTH parents are properly initialized
# This is the standard pattern for multiple inheritance with __init__
```

## Exercise 6: Online Ordering System (Abstraction)

```python
# PROBLEM:
# Create an abstract class Order with process_order() as an abstract method
# Create two child classes: OnlineOrder and StoreOrder
# Each must implement process_order() in their own way

# What this tests:
#   - Simple abstraction pattern
#   - Two different implementations of the same interface
#   - Foundation for real e-commerce systems
```

```python
# SOLUTION:

from abc import ABC, abstractmethod

class Order(ABC):
    @abstractmethod
    def process_order(self):
        pass

class OnlineOrder(Order):
    def process_order(self):
        print('Processing online order — shipping to customer address')

class StoreOrder(Order):
    def process_order(self):
        print('Processing store order — ready for pickup at counter')
```

```python
online = OnlineOrder()
store = StoreOrder()

online.process_order()    # Processing online order — shipping to customer address
store.process_order()     # Processing store order — ready for pickup at counter

# Same interface (process_order), completely different fulfilment logic
# This is why abstraction is powerful:
# The calling code doesn't need to know IF it's online or store
# It just calls process_order() and the right thing happens
```

## Exercise 7: Encapsulation — Constructor + Setter

```python
# PROBLEM:
# Create a class named Encapsulation:
#   a) Constructor initializes originalValue = 10
#   b) Method getValue() returns the originalValue
#   c) Method setValue(newValue) replaces originalValue with newValue

# What this tests:
#   - Constructor (__init__) to set initial values
#   - Getter method to READ the value
#   - Setter method to UPDATE the value
#   - This is the ENCAPSULATION pattern: control access to data through methods
```

```python
# SOLUTION:

class Encapsulation:
    def __init__(self):
        self.originalValue = 10     # initialized with a default value

    def getValue(self):             # getter — returns the current value
        return self.originalValue

    def setValue(self, newValue):    # setter — replaces the value
        self.originalValue = newValue
```

```python
obj = Encapsulation()

print('Original value:', obj.getValue())    # Original value: 10

obj.setValue(42)                             # change the value to 42

print('New value:', obj.getValue())          # New value: 42

# WHY use getter/setter instead of accessing obj.originalValue directly?
#
# In real code, you'd make originalValue PRIVATE (__originalValue)
# Then the only way to read or change it is through getValue() and setValue()
# This lets you add VALIDATION in the setter:
#   def setValue(self, newValue):
#       if newValue < 0:
#           print('Error: value cannot be negative')
#       else:
#           self.__originalValue = newValue
```

## Exercise 8: Factorial & Palindrome Functions

```python
# These were also part of the class handson (not OOP, but good practice)

# PROBLEM A: Create a function 'factor' that takes 1 number and returns its factorial
# Factorial: 5! = 5 × 4 × 3 × 2 × 1 = 120

def factor(n):
    result = 1
    for i in range(1, n + 1):    # 1 to n (inclusive)
        result = result * i
    return result

print('Factorial of 5:', factor(5))     # 120
print('Factorial of 7:', factor(7))     # 5040
print('Factorial of 0:', factor(0))     # 1 (by convention)
```

```python
# PROBLEM B: Create a function 'check_string' that takes a string
# and checks if it contains the letter 's'

def check_string(text):
    if 's' in text or 'S' in text:          # check both lowercase and uppercase
        print(f"The string '{text}' contains the letter 's'")
    else:
        print(f"The string '{text}' does NOT contain the letter 's'")

check_string('Python is awesome')    # contains 's'
check_string('Hello World')          # does NOT contain 's'
check_string('DevOps')               # contains 's'
```

```python
# PROBLEM C: Check if a string is a palindrome
# Palindrome: reads the same forwards and backwards
# Examples: 'madam', 'racecar' -> palindrome
#           'python' -> not a palindrome

def is_palindrome(text):
    text = text.lower()              # make case-insensitive
    if text == text[::-1]:           # compare with reverse
        print(f"'{text}' is a palindrome")
    else:
        print(f"'{text}' is NOT a palindrome")

is_palindrome('madam')       # 'madam' is a palindrome
is_palindrome('Racecar')     # 'racecar' is a palindrome
is_palindrome('Python')      # 'python' is NOT a palindrome
```

## Exercise 9: Lambda + *args

```python
# PROBLEM A: Create a lambda function that doubles a number

double_num = lambda x: x * 2        # lambda: one-line function

print(double_num(5))     # 10
print(double_num(12))    # 24
print(double_num(0))     # 0

# Equivalent to:
# def double_num(x):
#     return x * 2
# But lambda is more concise for simple operations
```

```python
# PROBLEM B: Create a function Sum that accepts multiple numbers using *args
# Add all numbers using a for loop and return the sum

def Sum(*args):                    # *args accepts any number of arguments
    total = 0
    for num in args:
        total += num
    return total

print('Sum:', Sum(10, 20, 30))            # Sum: 60
print('Sum:', Sum(5, 10, 15, 20, 25))     # Sum: 75
print('Sum:', Sum(100))                    # Sum: 100

# *args turns all the arguments into a TUPLE
# We loop through the tuple and add each number
# This is exactly what Python's built-in sum() does
```

## Exercise 10: Method Overloading (Why It Doesn't Work in Python)

```python
# PROBLEM:
# Create 2 methods named 'Hello'
# First Hello: takes 1 argument -> 'This function has 1 argument'
# Second Hello: takes 2 arguments -> 'This function has 2 arguments'
# Call both and observe what happens

# What this tests:
#   - Why method overloading does NOT work in Python
#   - The second definition REPLACES the first

class Demo:
    def Hello(self, a):
        print(f'This function has 1 argument: {a}')

    def Hello(self, a, b):    # This REPLACES the Hello(a) above!
        print(f'This function has 2 arguments: {a}, {b}')

d = Demo()
# d.Hello('hi')             # TypeError: Hello() missing 1 required positional argument: 'b'
d.Hello('hi', 'there')      # This function has 2 arguments: hi, there

# The 1-argument Hello is GONE — Python only keeps the LAST definition
# This is different from Java/C++ where overloading is supported
# In Python, use default parameters or *args instead
```

```python
# THE PYTHON WAY: use default parameters to handle variable arguments

class DemoBetter:
    def Hello(self, a, b=None):          # b is optional (defaults to None)
        if b is None:
            print(f'This function has 1 argument: {a}')
        else:
            print(f'This function has 2 arguments: {a}, {b}')

d = DemoBetter()
d.Hello('hi')               # This function has 1 argument: hi
d.Hello('hi', 'there')      # This function has 2 arguments: hi, there

# Now it works for BOTH cases!
```
