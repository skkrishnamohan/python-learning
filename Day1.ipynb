# Python - Introduction

```python
# Python - general purpose, high-level, interpreted programming language
# Created by: Guido van Rossum
# First released: 1991
# Named after: Monty Python's Flying Circus (British comedy show) - not the snake!

# Python's design philosophy -> code should be READABLE and SIMPLE
# "There should be one obvious way to do it"

# Current version: Python 3.x  (Python 2 retired in 2020 - don't use it!)
# File extension: .py

# Python is used by: Google, Netflix, Instagram, Dropbox, NASA, and basically every tech company
```

```python
# How code runs: Interpreted vs Compiled

# Compiled languages (C, C++, Go, Rust):
#   Source code  ->  COMPILER  ->  Machine code (binary)  ->  Execute
#   Advantages: Very fast execution
#   Disadvantage: must compile before running, platform-specific

# Interpreted languages (Python, JavaScript, Ruby):
#   Source code  ->  INTERPRETER reads & executes line by line
#   Advantages: No compile step, runs anywhere, easier to debug
#   Disadvantage: slightly slower than compiled (rarely matters in practice)

# Python is INTERPRETED -> write code and run it immediately (no compile step!)
# This is why Python is perfect for scripting and quick automation in DevOps

# NOTE: Internally Python does compile to bytecode (.pyc files automatically)
#       but this is transparent - we don't need to worry about it
```

```python
# Why Python is dominant in 2024+

# 1. Readable syntax  -> code reads almost like English, fast to learn
# 2. Huge ecosystem   -> 400,000+ packages on PyPI (pip install almost anything)
# 3. Versatile        -> web, AI/ML, data science, automation, networking, scripting
# 4. Cross-platform   -> same script runs on Windows, Linux, macOS unchanged
# 5. Strong community -> massive documentation, Stack Overflow, GitHub repos
# 6. Fast development -> write working code faster than Java or C++
# 7. Default for AI   -> PyTorch, TensorFlow, scikit-learn are all Python-first

# Python IS the default language for:
# - AI / Machine Learning  (numpy, pandas, scikit-learn, PyTorch, TensorFlow)
# - DevOps / Automation    (azure-sdk, paramiko, fabric, ansible uses Python)
# - Data Engineering       (PySpark, Apache Airflow, dbt - all Python)
# - Cloud SDKs             (Azure has first-class Python SDK support - azure-identity, azure-mgmt-*)
```

```python
# How to install and run Python

# ---- Installing ----
# 1. Download from https://www.python.org/downloads/  (latest Python 3.x)
# 2. During install: CHECK 'Add Python to PATH'  <- critical on Windows!
# 3. Verify: open terminal -> type:  python --version   or   python3 --version

# ---- On Linux/Mac servers (common DevOps scenario) ----
# Python 3 is usually pre-installed on most Linux distros
# Check:  python3 --version
# Install pip (package manager):
#   Ubuntu/Debian:  sudo apt install python3-pip
#   RHEL/CentOS:    sudo yum install python3-pip

# ---- Ways to run Python ----
# 1. REPL (interactive shell)  -> type 'python3' in terminal, type code live
# 2. Script file (.py)         -> python3 deploy_script.py  (most common in DevOps!)
# 3. Jupyter Notebook (.ipynb) -> what we're using - great for learning and ML work
# 4. IDE (VS Code, PyCharm)    -> for larger projects

# ---- Virtual Environments (IMPORTANT - learn this early!) ----
# Like a Docker container but for Python dependencies
# python3 -m venv myenv           -> create isolated environment
# source myenv/bin/activate        -> activate (Linux/Mac)
# myenv\Scripts\activate           -> activate (Windows)
# pip install <package>            -> installs into THIS env only
# deactivate                       -> exit the environment
# Always use venv for projects to avoid dependency conflicts!
```

# print() Function

```python
# print() is the function to print the text
#Anything within ('') will be consider as text
print('Hellow world')  #shift + Enter key to run
print('10-20')
```

```python
#For number no need to specify with ''
print(100)
print(10-20)
```

```python
print(10)
print(20)

#end parameter within print()
# if you haven't mentioned anything by default end='\n'
# end paramter controls what gets printed after the output

#when we simply do print function by default it will end value as /n that is
print(10,end='\n')
```

```python
#another example
print(10,end='-')
print(20)
```

```python
# sep parameter - controls what is printed BETWEEN multiple values
# by default sep is a space ' '

print('Python', 'is', 'fun')             # default sep = space
print('Python', 'is', 'fun', sep='-')    # sep = '-'
print('Python', 'is', 'fun', sep='')     # sep = nothing (no space)
print(2025, 6, 10, sep='/')              # useful for dates: 2025/6/10
```

```python
# combining end and sep together
print('Name:', 'Krishna', sep=' ', end=' | ')
print('Age:', 21)

# print() with no arguments -> just prints a blank line (useful as spacer)
print()
print('above is a blank line')
```
