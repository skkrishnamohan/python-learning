# Abstraction

```python
# Abstraction - showing ONLY the essential features, hiding unnecessary details

# Real-world analogy:
# When you press the brake in a car, you don't need to know HOW the braking mechanism works
# You just need to know THAT pressing brake slows down the car
# The internal complexity is HIDDEN from you

# Another analogy:
# ATM machine - you insert card, enter PIN, withdraw cash
# You don't know HOW the ATM communicates with the bank server internally
# That complexity is abstracted away from you

# In Python, abstraction is achieved using:
# 1. The 'abc' module (Abstract Base Classes)
# 2. @abstractmethod decorator

# An abstract class:
# - Cannot be instantiated (cannot create objects directly from it)
# - Acts as a BLUEPRINT - defines WHAT methods must exist
# - Does NOT define HOW those methods work (that's up to the child class)
# - Forces all child classes to implement its abstract methods
```

### Abstract Base Class (ABC) and @abstractmethod

```python
# Importing ABC (Abstract Base Class) and abstractmethod

from abc import ABC, abstractmethod

# Abstract class - a blueprint that CANNOT be instantiated directly
class Animal(ABC):

    @abstractmethod
    def sound(self):    # abstract method - NO body, child MUST define it
        pass

    def breathe(self):  # concrete method - has a body, inherited as-is
        print('All animals breathe oxygen')

# Child class MUST implement the abstract method
class Dog(Animal):
    def sound(self):
        print('Dog barks')

class Cat(Animal):
    def sound(self):
        print('Cat meows')

d = Dog()
c = Cat()

d.sound()      # Dog barks
d.breathe()    # All animals breathe oxygen
c.sound()      # Cat meows
```

```python
# Key points about abstract class:

# 1. It can have BOTH abstract methods AND concrete (normal) methods
#    - abstract method -> no body, child MUST override it
#    - concrete method -> has a body, child inherits it as-is (or can override)

# 2. A class MUST inherit from ABC to be treated as abstract
#    - from abc import ABC, abstractmethod
#    - class MyClass(ABC):   <-- this makes it abstract

# 3. @abstractmethod decorator marks a method as abstract
#    - forces child classes to provide their own implementation
```

### Cannot Instantiate an Abstract Class

```python
# Trying to create an object of the abstract class directly -> TypeError

from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass

a = Animal()    # TypeError: Can't instantiate abstract class Animal with abstract method sound

# WHY? Because the abstract class is just a BLUEPRINT
# It says 'any Animal must have sound()' but doesn't define what the sound IS
# So creating a generic 'Animal' object doesn't make sense
```

### Child MUST Implement All Abstract Methods

```python
# If a child class does NOT implement all abstract methods -> also TypeError

from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass

class Fish(Animal):     # Fish does NOT implement sound()
    def swim(self):
        print('Fish swims')

f = Fish()    # TypeError: Can't instantiate abstract class Fish with abstract method sound

# The RULE: if your parent has abstract methods, you MUST implement ALL of them
# Otherwise Python won't let you create an object
```

### Abstract Class with Multiple Abstract Methods

```python
# An abstract class can have MULTIPLE abstract methods
# The child MUST implement ALL of them

from abc import ABC, abstractmethod

class Vehicle(ABC):
    @abstractmethod
    def start(self):
        pass

    @abstractmethod
    def stop(self):
        pass

    @abstractmethod
    def fuel_type(self):
        pass

    def description(self):       # concrete method - inherited as-is
        print('This is a vehicle')

class Car(Vehicle):
    def start(self):
        print('Car starts with key ignition')

    def stop(self):
        print('Car stops with brake pedal')

    def fuel_type(self):
        print('Car runs on petrol/diesel')

class ElectricScooter(Vehicle):
    def start(self):
        print('Scooter starts with a button press')

    def stop(self):
        print('Scooter stops with hand brake')

    def fuel_type(self):
        print('Scooter runs on electric battery')

c = Car()
e = ElectricScooter()

c.start()          # Car starts with key ignition
c.fuel_type()      # Car runs on petrol/diesel
c.description()    # This is a vehicle (inherited)

e.start()          # Scooter starts with a button press
e.fuel_type()      # Scooter runs on electric battery
```

### Real-World Example: Payment System

```python
# Scenario: Payment system using abstraction
# Different payment methods (UPI, Card, Cash) all MUST implement pay()

from abc import ABC, abstractmethod

class Payment(ABC):        # abstract class
    @abstractmethod
    def pay(self, amount): # every payment method must implement this
        pass

    @abstractmethod
    def refund(self, amount):  # every payment method must handle refunds too
        pass

class UPI(Payment):
    def pay(self, amount):
        print('Paid Rs', amount, 'via UPI')

    def refund(self, amount):
        print('Refunded Rs', amount, 'to UPI wallet')

class Card(Payment):
    def pay(self, amount):
        print('Paid Rs', amount, 'via Credit Card')

    def refund(self, amount):
        print('Refunded Rs', amount, 'to Credit Card')

class Cash(Payment):
    def pay(self, amount):
        print('Paid Rs', amount, 'via Cash')

    def refund(self, amount):
        print('Cash refund of Rs', amount, 'given')

u = UPI()
k = Card()
c = Cash()

u.pay(500)
k.pay(1200)
c.pay(200)

print()  # blank line

u.refund(500)
k.refund(1200)
```

### Abstraction with __init__ in Abstract Class

```python
# Abstract classes CAN have __init__ method
# Child classes use super().__init__() to call it

from abc import ABC, abstractmethod

class Employee(ABC):
    def __init__(self, name, empid):
        self.name  = name
        self.empid = empid

    @abstractmethod
    def calculate_salary(self):
        pass

    def display(self):    # concrete method
        print(f'Name: {self.name}, EmpID: {self.empid}')

class FullTime(Employee):
    def __init__(self, name, empid, monthly_salary):
        super().__init__(name, empid)
        self.monthly_salary = monthly_salary

    def calculate_salary(self):
        print(f'{self.name} earns Rs {self.monthly_salary} per month (fixed)')

class PartTime(Employee):
    def __init__(self, name, empid, hours_worked, hourly_rate):
        super().__init__(name, empid)
        self.hours_worked = hours_worked
        self.hourly_rate  = hourly_rate

    def calculate_salary(self):
        total = self.hours_worked * self.hourly_rate
        print(f'{self.name} earns Rs {total} ({self.hours_worked} hrs x Rs {self.hourly_rate}/hr)')

f = FullTime('Rahul', 1001, 80000)
p = PartTime('Priya', 1002, 120, 500)

f.display()             # Name: Rahul, EmpID: 1001
f.calculate_salary()    # Rahul earns Rs 80000 per month (fixed)

print()

p.display()             # Name: Priya, EmpID: 1002
p.calculate_salary()    # Priya earns Rs 60000 (120 hrs x Rs 500/hr)
```

### Abstraction vs Encapsulation — What's the Difference?

```python
# People often confuse Abstraction and Encapsulation. Here's the difference:

# ABSTRACTION:
# - Hides IMPLEMENTATION details (the HOW)
# - Focuses on WHAT an object does, not HOW it does it
# - Achieved using: abstract classes (ABC + @abstractmethod)
# - Example: 'Payment must have a pay() method' — but each payment type implements it differently

# ENCAPSULATION:
# - Hides DATA (the internal state)
# - Restricts DIRECT ACCESS to an object's attributes
# - Achieved using: private attributes (__name), getters/setters
# - Example: 'Salary is private, you can only read it through get_salary()'

# Simple way to remember:
# Abstraction  = hiding the COMPLEXITY (what methods exist vs how they work)
# Encapsulation = hiding the DATA (restricting access to variables)
```

### Quick Recap — All 4 OOP Pillars

```python
# Quick recap of all 4 OOP pillars:

# 1. Encapsulation  - bundling data + methods, using access modifiers to restrict access
#                     Tools: public, _protected, __private, getter/setter

# 2. Inheritance    - child class reuses parent class code
#                     Tools: class Child(Parent), super()

# 3. Polymorphism   - same method name, different behaviour per object
#                     Tools: method overriding in child classes

# 4. Abstraction    - hide implementation details, expose only essential interface
#                     Tools: ABC, @abstractmethod

# Now you know ALL 4 pillars of OOP!
# These 4 concepts are the FOUNDATION of object-oriented design
# Every framework you'll use (Django, Flask, FastAPI, LangChain) uses these patterns
```
